```python
# добавим категориальные переменные по среднему количеству населения
def get_population_cat(mark):     
    if mark <= 1000000: 
        return 'Очень малочисленная' 
    elif  375000000 >= mark > 1000000 : 
        return 'Малочисленная' 
    elif 750000000 >= mark > 375000000 : 
        return 'Средняя' 
    elif 1125000000 >= mark > 750000000 : 
        return 'Высокая' 
    elif mark > 1400000000: 
        return 'Очень высокая' 
```

```python
#категориальные переменные для колонки 2022
world_population['Категория популяции'] = world_population['2022'].apply(get_population_cat)  
```

Также добавим категориальные переменные для определения того, как выросла популяция стран с 1970 по 2022. Это поможет нам определить, сильно ли за этот промежуток времени выросла популяция в каждой стране.

```python
world_population.head(10)
```

```python
# добавим категориальные переменные по среднему количеству населения
def get_population_cat(mark):     
    if mark <= 1000000: 
        return 'Очень мало' 
    elif  5000000 >= mark > 1000000 : 
        return 'Мало' 
    elif 10000000 >= mark > 5000000 : 
        return 'Умеренно' 
    elif 300000000 >= mark > 10000000 : 
        return 'Сильно' 
    elif mark > 1000000000: 
        return 'Очень сильно' 
```

```python
#категориальные переменные для колонки Рост населения с 1970 по 2022
world_population['Увеличение населения'] = world_population['Рост населения 1970-2022'].apply(get_population_cat)   
```
